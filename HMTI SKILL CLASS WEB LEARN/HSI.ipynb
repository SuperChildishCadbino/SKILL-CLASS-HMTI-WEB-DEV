```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

# Definisikan titik-titik
points = [
    [(6, 5, 8), (8, -7, -3)],
    [(1, 2, 3), (4, 5, 6)],
    [(1, 4, 4), (2, -4, 6)]
]

# Buat plot 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Loop melalui setiap pasangan titik dan plot vektornya
for p1, p2 in points:
    # Hitung komponen vektor
    v = np.array(p2) - np.array(p1)
    
    # Plot titik awal
    ax.scatter(p1[0], p1[1], p1[2], color='r')
    ax.text(p1[0], p1[1], p1[2], f'({p1[0]},{p1[1]},{p1[2]})')
    
    # Plot vektor
    ax.quiver(p1[0], p1[1], p1[2], v[0], v[1], v[2], color='b')
    ax.text(p1[0] + v[0], p1[1] + v[1], p1[2] + v[2], f'({v[0]},{v[1]},{v[2]})')

# Tentukan batas sumbu
ax.set_xlim([0, 10])
ax.set_ylim([-10, 10])
ax.set_zlim([-10, 10])

# Label sumbu
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

plt.title('Plot Vektor')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Definisikan vektor v1
v1 = (2, 5)

# Mendapatkan komponen x dan y dari vektor v1
x, y = v1

# Membuat plot
plt.figure()
plt.quiver(0, 0, x, y, angles='xy', scale_units='xy', scale=1, color='b', label='v1')

# Mengatur batas sumbu x dan y
plt.xlim(-6, 7)
plt.ylim(-1, 8)

# Menandai sumbu x dan y
plt.axhline(0, color='k', linewidth=0.5)
plt.axvline(0, color='k', linewidth=0.5)

# Menandai label sumbu x dan y
plt.xlabel('X')
plt.ylabel('Y')

# Menambahkan legenda
plt.legend()

# Menampilkan grid
plt.grid(True, linestyle='--', alpha=0.5)

# Menampilkan plot
plt.title('Grafik Vektor v1')
plt.show()
```
